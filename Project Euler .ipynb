```python
def isPalindrome(num_1):
    """
    CONVERTS THE NUMBER TO A STRING AND CHECKS IF THE SAME IN REVERSE; RETURNS A BOOLENS RESULT 
    """
    return str(num_1) == str(num_1)[::-1]
```

```python
def largestPalindrome(lower_limit,upper_limit):
    """
    CHECKS ALL THE NUMBES FROM LOWER_LIMIT TO UPPER LIMIT AND RETURNS THE LARGEST PALINDROME
    """
    z = 0;
    for x in range(lower_limit,upper_limit):
        for y in range(lower_limit,upper_limit):
            result = x * y
            if isPalindrome(result):
                if result > z:
                    z = result 
    return(z)
```

```python
from math import sqrt

def divisors(n):
    divisor_list = []
    
    limit = int(sqrt(n))
    for i in range(1,limit+1):

        if n % i == 0:

            divisor_list.append(i)

            if i != n/i:
                divisor_list.append(n/i)
    
    return len(divisor_list)  

```

```python
def get_power_of_four(n):
    num_list = []
    for num in range (2,n):
        number = sum([int(i)**4 for i in str(num)])
        
        if num == number :
            print(num)
            num_list.append(num)
    return num_list
        
        
```

```python
def isPalindrome(num_1):
    """
    CONVERTS THE NUMBER TO A STRING AND CHECKS IF THE SAME IN REVERSE; RETURNS A BOOLENS RESULT 
    """
    return str(num_1) == str(num_1)[::-1]
```

```python
pal_list = []
for i in range(1,1000000):
    if isPalindrome(i):
        if isPalindrome(int(bin(i)[2:])):
#             print(i)
            pal_list.append(i)
            
sum(pal_list)            
            
```
